**Pre Processing the Data**

```python
nltk.download('stopwords')
lemmatizer = WordNetLemmatizer()
stpwrds = list(stopwords.words('english'))
```

```python
stpwrds
```

**LEMMATIZATION**

```python
#lemmatization
nltk.download('wordnet')
nltk.download('punkt')
for x in range(len(train_df)) :
    corpus = []
    review = train_df['text'][x]
    review = re.sub(r'[^a-zA-Z\s]', '', review)
    review = review.lower()
    review = nltk.word_tokenize(review)
    for y in review :
        if y not in stpwrds :
            corpus.append(lemmatizer.lemmatize(y))
    review = ' '.join(corpus)
    train_df['text'][x] = review
```

```python
train_df['text'][2182]
```

**SPLITTING THE DATASET INTO TRAINING DATA & TEST DATA:**

```python
X_train, X_test, Y_train, Y_test = train_test_split(train_df['text'], label_train, test_size=0.2, random_state=0)
```

```python
X_train.shape
```

**Converting Textual Data into Numerical Data using TFIDF vectorizer.**

```python
#converting to textual data into numerical data/feature vector

tfidf_vect = TfidfVectorizer()
tfidf_X_train = tfidf_vect.fit_transform(X_train)
tfidf_X_test = tfidf_vect.transform(X_test)
```

```python
print(tfidf_X_train,tfidf_X_test)
```

```python
tfidf_X_train.shape
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

**Model Training**

```python
classifier = PassiveAggressiveClassifier()
classifier.fit(tfidf_X_train,Y_train)
```

**ACCURACY VALUE WITH CONFUSION MATRIX**

```python
#accuracy score on the test data
Y_pred = classifier.predict(tfidf_X_test)
score = metrics.accuracy_score(Y_test, Y_pred)
print(f'Accuracy of the data: {round(score*100,2)}%')
cm = metrics.confusion_matrix(Y_test, Y_pred)
plot_confusion_matrix(cm, classes=['FAKE DATA', 'REAL DATA'])
```

```python
print(classification_report(Y_test, Y_pred))
```

```python
#pickling the model into disk
pickle.dump(classifier,open('./model.pkl', 'wb'))
pickle.dump(tfidf_vect,open('vector.pkl', 'wb'))
```

```python
# load the model from disk
loaded_model = pickle.load(open('./model.pkl', 'rb'))
#vector_form = pickle.load(open('vector.pkl', 'rb')
```

**Prediction of data**

```python
def fake_news_det(news):
    review = news
    review = re.sub(r'[^a-zA-Z\s]', '', review)
    review = review.lower()
    review = nltk.word_tokenize(review)
    corpus = []
    for y in review :
        if y not in stpwrds :
            corpus.append(lemmatizer.lemmatize(y))
    input_data = [' '.join(corpus)]
    vectorized_input_data = tfidf_vect.transform(input_data)
    prediction = loaded_model.predict(vectorized_input_data)
    if prediction[0] == 1:
        print("Prediction of the News :  Looking Fake⚠ News📰 ")
    else:
        print("Prediction of the News : Looking Real News📰 ")
```
